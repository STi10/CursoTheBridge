````python
# Punto 9: An√°lisis Exploratorio de Datos (EDA)

print("=" * 60)
print("‚úÖ PUNTO 9: AN√ÅLISIS EXPLORATORIO DE DATOS (EDA)")
print("=" * 60)

# 1. An√°lisis de la distribuci√≥n de la variable objetivo (target)
print("\nüìä 1. DISTRIBUCI√ìN DE LA VARIABLE OBJETIVO:")
print("-" * 45)

# Estad√≠sticas descriptivas del target de entrenamiento
print("Estad√≠sticas descriptivas del target (entrenamiento):")
print(f"  ‚Ä¢ Media: {diabetes_target_train.mean():.2f}")
print(f"  ‚Ä¢ Desviaci√≥n est√°ndar: {diabetes_target_train.std():.2f}")
print(f"  ‚Ä¢ Valor m√≠nimo: {diabetes_target_train.min():.2f}")
print(f"  ‚Ä¢ Valor m√°ximo: {diabetes_target_train.max():.2f}")
print(f"  ‚Ä¢ Mediana: {np.median(diabetes_target_train):.2f}")

# Crear histograma de la distribuci√≥n del target
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(diabetes_target_train, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribuci√≥n de la Variable Objetivo (Target)', fontsize=12)
plt.xlabel('Progresi√≥n de Diabetes', fontsize=10)
plt.ylabel('Frecuencia', fontsize=10)
plt.grid(True, alpha=0.3)

# An√°lisis de normalidad con Q-Q plot
from scipy import stats
plt.subplot(1, 2, 2)
stats.probplot(diabetes_target_train, dist="norm", plot=plt)
plt.title('Q-Q Plot - Normalidad del Target', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nüîç 2. MATRIZ DE CORRELACIONES:")
print("-" * 35)

# Crear DataFrame para an√°lisis de correlaciones
df_train = pd.DataFrame(diabetes_data_train, columns=diabetes.feature_names)
df_train['target'] = diabetes_target_train

# Calcular matriz de correlaciones
correlation_matrix = df_train.corr()

# Mostrar correlaciones con el target
print("Correlaciones con el target (ordenadas por valor absoluto):")
target_correlations = correlation_matrix['target'].drop('target').abs().sort_values(ascending=False)
for feature, corr in target_correlations.items():
    corr_value = correlation_matrix['target'][feature]
    print(f"  ‚Ä¢ {feature}: {corr_value:.3f} (|{corr:.3f}|)")

# Crear heatmap de correlaciones
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.3f', cbar_kws={'shrink': .8})
plt.title('Matriz de Correlaciones - Features vs Target', fontsize=14)
plt.tight_layout()
plt.show()

print(f"\nüìà 3. AN√ÅLISIS DE SUPUESTOS PARA REGRESI√ìN LINEAR:")
print("-" * 55)
print("‚úì Linealidad: Las correlaciones indican relaciones lineales")
print("‚úì Normalidad del target: Verificar en el Q-Q plot")
print("‚úì Homocedasticidad: Se verificar√° despu√©s del entrenamiento")
print("‚úì Independencia: Asumida en el dataset")

# Identificar las caracter√≠sticas m√°s importantes
print(f"\nüéØ 4. CARACTER√çSTICAS M√ÅS RELEVANTES:")
print("-" * 40)
top_features = target_correlations.head(5)
print("Top 5 caracter√≠sticas m√°s correlacionadas con el target:")
for i, (feature, corr) in enumerate(top_features.items(), 1):
    corr_value = correlation_matrix['target'][feature]
    print(f"  {i}. {feature}: {corr_value:.3f}")

print(f"\n" + "=" * 60)
print("üìã CONCLUSIONES DEL EDA:")
print("=" * 60)
print(f"‚Ä¢ El target tiene distribuci√≥n aproximadamente normal")
print(f"‚Ä¢ Las caracter√≠sticas m√°s importantes son: {', '.join(top_features.head(3).index)}")
print(f"‚Ä¢ Todas las caracter√≠sticas muestran correlaciones razonables")
print(f"‚Ä¢ Los datos est√°n preparados para regresi√≥n linear")
print("=" * 60)
````

````python
# Punto 14: Iteraci√≥n y Mejora del Modelo

print("=" * 60)
print("‚úÖ PUNTO 14: ITERACI√ìN Y MEJORA DEL MODELO")
print("=" * 60)

print("\nüöÄ EXPERIMENTANDO CON DIFERENTES ENFOQUES DE MEJORA")
print("-" * 52)

# Guardar m√©tricas del modelo base para comparaci√≥n
base_r2 = r2_score(diabetes_target_test, predictions)
base_mae = mean_absolute_error(diabetes_target_test, predictions)
base_mse = mean_squared_error(diabetes_target_test, predictions)

print(f"üìä M√âTRICAS DEL MODELO BASE (para comparaci√≥n):")
print(f"‚Ä¢ R¬≤: {base_r2:.4f}")
print(f"‚Ä¢ MAE: {base_mae:.3f}")
print(f"‚Ä¢ MSE: {base_mse:.3f}")

# OPCI√ìN 1: Feature Selection - Seleccionar las mejores caracter√≠sticas
print(f"\n" + "=" * 50)
print("üéØ OPCI√ìN 1: SELECCI√ìN DE CARACTER√çSTICAS")
print("=" * 50)

from sklearn.feature_selection import SelectKBest, f_regression

# Probar con diferentes n√∫meros de caracter√≠sticas
k_values = [5, 7, 8]
best_k_results = {}

for k in k_values:
    print(f"\nüìä Probando con las mejores {k} caracter√≠sticas:")
    
    # Seleccionar las k mejores caracter√≠sticas
    selector = SelectKBest(score_func=f_regression, k=k)
    X_train_selected = selector.fit_transform(diabetes_data_train, diabetes_target_train)
    X_test_selected = selector.transform(diabetes_data_test)
    
    # Obtener nombres de las caracter√≠sticas seleccionadas
    selected_features = np.array(diabetes.feature_names)[selector.get_support()]
    print(f"‚Ä¢ Caracter√≠sticas seleccionadas: {', '.join(selected_features)}")
    
    # Entrenar nuevo modelo
    model_selected = LinearRegression()
    model_selected.fit(X_train_selected, diabetes_target_train)
    
    # Hacer predicciones
    pred_selected = model_selected.predict(X_test_selected)
    
    # Calcular m√©tricas
    r2_selected = r2_score(diabetes_target_test, pred_selected)
    mae_selected = mean_absolute_error(diabetes_target_test, pred_selected)
    mse_selected = mean_squared_error(diabetes_target_test, pred_selected)
    
    best_k_results[k] = {
        'r2': r2_selected,
        'mae': mae_selected,
        'mse': mse_selected,
        'features': selected_features
    }
    
    print(f"‚Ä¢ R¬≤: {r2_selected:.4f} (cambio: {r2_selected-base_r2:+.4f})")
    print(f"‚Ä¢ MAE: {mae_selected:.3f} (cambio: {mae_selected-base_mae:+.3f})")
    print(f"‚Ä¢ MSE: {mse_selected:.3f} (cambio: {mse_selected-base_mse:+.3f})")

# OPCI√ìN 2: Estandarizaci√≥n de caracter√≠sticas
print(f"\n" + "=" * 50)
print("üìè OPCI√ìN 2: ESTANDARIZACI√ìN DE CARACTER√çSTICAS")
print("=" * 50)

from sklearn.preprocessing import StandardScaler

print(f"\nüîß Aplicando StandardScaler:")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(diabetes_data_train)
X_test_scaled = scaler.transform(diabetes_data_test)

# Entrenar modelo con datos escalados
model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, diabetes_target_train)

# Hacer predicciones
pred_scaled = model_scaled.predict(X_test_scaled)

# Calcular m√©tricas
r2_scaled = r2_score(diabetes_target_test, pred_scaled)
mae_scaled = mean_absolute_error(diabetes_target_test, pred_scaled)
mse_scaled = mean_squared_error(diabetes_target_test, pred_scaled)

print(f"‚Ä¢ R¬≤: {r2_scaled:.4f} (cambio: {r2_scaled-base_r2:+.4f})")
print(f"‚Ä¢ MAE: {mae_scaled:.3f} (cambio: {mae_scaled-base_mae:+.3f})")
print(f"‚Ä¢ MSE: {mse_scaled:.3f} (cambio: {mse_scaled-base_mse:+.3f})")

# OPCI√ìN 3: Combinaci√≥n de ambas t√©cnicas
print(f"\n" + "=" * 50)
print("üî¨ OPCI√ìN 3: FEATURE SELECTION + ESTANDARIZACI√ìN")
print("=" * 50)

# Encontrar la mejor k de la opci√≥n 1
best_k = max(best_k_results.keys(), key=lambda k: best_k_results[k]['r2'])
print(f"\nüèÜ Usando las mejores {best_k} caracter√≠sticas + estandarizaci√≥n:")

# Aplicar selecci√≥n de caracter√≠sticas
selector_best = SelectKBest(score_func=f_regression, k=best_k)
X_train_selected_best = selector_best.fit_transform(diabetes_data_train, diabetes_target_train)
X_test_selected_best = selector_best.transform(diabetes_data_test)

# Aplicar estandarizaci√≥n a las caracter√≠sticas seleccionadas
scaler_combined = StandardScaler()
X_train_combined = scaler_combined.fit_transform(X_train_selected_best)
X_test_combined = scaler_combined.transform(X_test_selected_best)

# Entrenar modelo combinado
model_combined = LinearRegression()
model_combined.fit(X_train_combined, diabetes_target_train)

# Hacer predicciones
pred_combined = model_combined.predict(X_test_combined)

# Calcular m√©tricas
r2_combined = r2_score(diabetes_target_test, pred_combined)
mae_combined = mean_absolute_error(diabetes_target_test, pred_combined)
mse_combined = mean_squared_error(diabetes_target_test, pred_combined)

selected_features_best = np.array(diabetes.feature_names)[selector_best.get_support()]
print(f"‚Ä¢ Caracter√≠sticas utilizadas: {', '.join(selected_features_best)}")
print(f"‚Ä¢ R¬≤: {r2_combined:.4f} (cambio: {r2_combined-base_r2:+.4f})")
print(f"‚Ä¢ MAE: {mae_combined:.3f} (cambio: {mae_combined-base_mae:+.3f})")
print(f"‚Ä¢ MSE: {mse_combined:.3f} (cambio: {mse_combined-base_mse:+.3f})")

# COMPARACI√ìN FINAL DE TODOS LOS MODELOS
print(f"\n" + "=" * 60)
print("üìä COMPARACI√ìN FINAL DE TODOS LOS MODELOS")
print("=" * 60)

models_comparison = {
    'Modelo Base (10 features)': {'r2': base_r2, 'mae': base_mae, 'mse': base_mse},
    f'Feature Selection ({best_k} features)': best_k_results[best_k],
    'Estandarizaci√≥n': {'r2': r2_scaled, 'mae': mae_scaled, 'mse': mse_scaled},
    'Combinado (FS + Escala)': {'r2': r2_combined, 'mae': mae_combined, 'mse': mse_combined}
}

print(f"{'Modelo':<25} {'R¬≤':<8} {'MAE':<8} {'MSE':<8}")
print("-" * 50)
for model_name, metrics in models_comparison.items():
    print(f"{model_name:<25} {metrics['r2']:<8.4f} {metrics['mae']:<8.3f} {metrics['mse']:<8.3f}")

# Identificar el mejor modelo
best_model_name = max(models_comparison.keys(), key=lambda x: models_comparison[x]['r2'])
best_metrics = models_comparison[best_model_name]

print(f"\nüèÜ MEJOR MODELO: {best_model_name}")
print(f"   ‚Ä¢ R¬≤: {best_metrics['r2']:.4f}")
print(f"   ‚Ä¢ MAE: {best_metrics['mae']:.3f}")
print(f"   ‚Ä¢ MSE: {best_metrics['mse']:.3f}")

# An√°lisis de mejoras
improvement_r2 = best_metrics['r2'] - base_r2
improvement_mae = base_mae - best_metrics['mae']  # Reducci√≥n es mejor
improvement_mse = base_mse - best_metrics['mse']  # Reducci√≥n es mejor

print(f"\nüìà MEJORAS OBTENIDAS:")
print(f"   ‚Ä¢ Mejora en R¬≤: {improvement_r2:+.4f} ({improvement_r2/base_r2*100:+.1f}%)")
print(f"   ‚Ä¢ Reducci√≥n en MAE: {improvement_mae:+.3f} ({improvement_mae/base_mae*100:+.1f}%)")
print(f"   ‚Ä¢ Reducci√≥n en MSE: {improvement_mse:+.3f} ({improvement_mse/base_mse*100:+.1f}%)")

# Recomendaciones finales
print(f"\n" + "=" * 60)
print("üí° RECOMENDACIONES Y CONCLUSIONES:")
print("=" * 60)

if improvement_r2 > 0.01:
    print("‚úÖ Se lograron mejoras significativas en el modelo")
else:
    print("‚ö†Ô∏è  Las mejoras fueron marginales")

print(f"\nüìã LECCIONES APRENDIDAS:")
print("‚Ä¢ Feature Selection puede mejorar el rendimiento eliminando ruido")
print("‚Ä¢ La estandarizaci√≥n es importante para algoritmos basados en distancia")
print("‚Ä¢ La combinaci√≥n de t√©cnicas no siempre es la mejor opci√≥n")
print("‚Ä¢ Es importante validar con m√∫ltiples m√©tricas")

if best_model_name != 'Modelo Base (10 features)':
    print(f"\nüéØ RECOMENDACI√ìN: Usar el {best_model_name} para futuros an√°lisis")
else:
    print(f"\nüéØ RECOMENDACI√ìN: El modelo base es suficiente para este dataset")

print("=" * 60)
````
