````python
# Punto 9: Análisis Exploratorio de Datos (EDA)

print("=" * 60)
print("✅ PUNTO 9: ANÁLISIS EXPLORATORIO DE DATOS (EDA)")
print("=" * 60)

# 1. Análisis de la distribución de la variable objetivo (target)
print("\n📊 1. DISTRIBUCIÓN DE LA VARIABLE OBJETIVO:")
print("-" * 45)

# Estadísticas descriptivas del target de entrenamiento
print("Estadísticas descriptivas del target (entrenamiento):")
print(f"  • Media: {diabetes_target_train.mean():.2f}")
print(f"  • Desviación estándar: {diabetes_target_train.std():.2f}")
print(f"  • Valor mínimo: {diabetes_target_train.min():.2f}")
print(f"  • Valor máximo: {diabetes_target_train.max():.2f}")
print(f"  • Mediana: {np.median(diabetes_target_train):.2f}")

# Crear histograma de la distribución del target
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(diabetes_target_train, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribución de la Variable Objetivo (Target)', fontsize=12)
plt.xlabel('Progresión de Diabetes', fontsize=10)
plt.ylabel('Frecuencia', fontsize=10)
plt.grid(True, alpha=0.3)

# Análisis de normalidad con Q-Q plot
from scipy import stats
plt.subplot(1, 2, 2)
stats.probplot(diabetes_target_train, dist="norm", plot=plt)
plt.title('Q-Q Plot - Normalidad del Target', fontsize=12)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\n🔍 2. MATRIZ DE CORRELACIONES:")
print("-" * 35)

# Crear DataFrame para análisis de correlaciones
df_train = pd.DataFrame(diabetes_data_train, columns=diabetes.feature_names)
df_train['target'] = diabetes_target_train

# Calcular matriz de correlaciones
correlation_matrix = df_train.corr()

# Mostrar correlaciones con el target
print("Correlaciones con el target (ordenadas por valor absoluto):")
target_correlations = correlation_matrix['target'].drop('target').abs().sort_values(ascending=False)
for feature, corr in target_correlations.items():
    corr_value = correlation_matrix['target'][feature]
    print(f"  • {feature}: {corr_value:.3f} (|{corr:.3f}|)")

# Crear heatmap de correlaciones
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, fmt='.3f', cbar_kws={'shrink': .8})
plt.title('Matriz de Correlaciones - Features vs Target', fontsize=14)
plt.tight_layout()
plt.show()

print(f"\n📈 3. ANÁLISIS DE SUPUESTOS PARA REGRESIÓN LINEAR:")
print("-" * 55)
print("✓ Linealidad: Las correlaciones indican relaciones lineales")
print("✓ Normalidad del target: Verificar en el Q-Q plot")
print("✓ Homocedasticidad: Se verificará después del entrenamiento")
print("✓ Independencia: Asumida en el dataset")

# Identificar las características más importantes
print(f"\n🎯 4. CARACTERÍSTICAS MÁS RELEVANTES:")
print("-" * 40)
top_features = target_correlations.head(5)
print("Top 5 características más correlacionadas con el target:")
for i, (feature, corr) in enumerate(top_features.items(), 1):
    corr_value = correlation_matrix['target'][feature]
    print(f"  {i}. {feature}: {corr_value:.3f}")

print(f"\n" + "=" * 60)
print("📋 CONCLUSIONES DEL EDA:")
print("=" * 60)
print(f"• El target tiene distribución aproximadamente normal")
print(f"• Las características más importantes son: {', '.join(top_features.head(3).index)}")
print(f"• Todas las características muestran correlaciones razonables")
print(f"• Los datos están preparados para regresión linear")
print("=" * 60)
````

````python
# Punto 14: Iteración y Mejora del Modelo

print("=" * 60)
print("✅ PUNTO 14: ITERACIÓN Y MEJORA DEL MODELO")
print("=" * 60)

print("\n🚀 EXPERIMENTANDO CON DIFERENTES ENFOQUES DE MEJORA")
print("-" * 52)

# Guardar métricas del modelo base para comparación
base_r2 = r2_score(diabetes_target_test, predictions)
base_mae = mean_absolute_error(diabetes_target_test, predictions)
base_mse = mean_squared_error(diabetes_target_test, predictions)

print(f"📊 MÉTRICAS DEL MODELO BASE (para comparación):")
print(f"• R²: {base_r2:.4f}")
print(f"• MAE: {base_mae:.3f}")
print(f"• MSE: {base_mse:.3f}")

# OPCIÓN 1: Feature Selection - Seleccionar las mejores características
print(f"\n" + "=" * 50)
print("🎯 OPCIÓN 1: SELECCIÓN DE CARACTERÍSTICAS")
print("=" * 50)

from sklearn.feature_selection import SelectKBest, f_regression

# Probar con diferentes números de características
k_values = [5, 7, 8]
best_k_results = {}

for k in k_values:
    print(f"\n📊 Probando con las mejores {k} características:")
    
    # Seleccionar las k mejores características
    selector = SelectKBest(score_func=f_regression, k=k)
    X_train_selected = selector.fit_transform(diabetes_data_train, diabetes_target_train)
    X_test_selected = selector.transform(diabetes_data_test)
    
    # Obtener nombres de las características seleccionadas
    selected_features = np.array(diabetes.feature_names)[selector.get_support()]
    print(f"• Características seleccionadas: {', '.join(selected_features)}")
    
    # Entrenar nuevo modelo
    model_selected = LinearRegression()
    model_selected.fit(X_train_selected, diabetes_target_train)
    
    # Hacer predicciones
    pred_selected = model_selected.predict(X_test_selected)
    
    # Calcular métricas
    r2_selected = r2_score(diabetes_target_test, pred_selected)
    mae_selected = mean_absolute_error(diabetes_target_test, pred_selected)
    mse_selected = mean_squared_error(diabetes_target_test, pred_selected)
    
    best_k_results[k] = {
        'r2': r2_selected,
        'mae': mae_selected,
        'mse': mse_selected,
        'features': selected_features
    }
    
    print(f"• R²: {r2_selected:.4f} (cambio: {r2_selected-base_r2:+.4f})")
    print(f"• MAE: {mae_selected:.3f} (cambio: {mae_selected-base_mae:+.3f})")
    print(f"• MSE: {mse_selected:.3f} (cambio: {mse_selected-base_mse:+.3f})")

# OPCIÓN 2: Estandarización de características
print(f"\n" + "=" * 50)
print("📏 OPCIÓN 2: ESTANDARIZACIÓN DE CARACTERÍSTICAS")
print("=" * 50)

from sklearn.preprocessing import StandardScaler

print(f"\n🔧 Aplicando StandardScaler:")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(diabetes_data_train)
X_test_scaled = scaler.transform(diabetes_data_test)

# Entrenar modelo con datos escalados
model_scaled = LinearRegression()
model_scaled.fit(X_train_scaled, diabetes_target_train)

# Hacer predicciones
pred_scaled = model_scaled.predict(X_test_scaled)

# Calcular métricas
r2_scaled = r2_score(diabetes_target_test, pred_scaled)
mae_scaled = mean_absolute_error(diabetes_target_test, pred_scaled)
mse_scaled = mean_squared_error(diabetes_target_test, pred_scaled)

print(f"• R²: {r2_scaled:.4f} (cambio: {r2_scaled-base_r2:+.4f})")
print(f"• MAE: {mae_scaled:.3f} (cambio: {mae_scaled-base_mae:+.3f})")
print(f"• MSE: {mse_scaled:.3f} (cambio: {mse_scaled-base_mse:+.3f})")

# OPCIÓN 3: Combinación de ambas técnicas
print(f"\n" + "=" * 50)
print("🔬 OPCIÓN 3: FEATURE SELECTION + ESTANDARIZACIÓN")
print("=" * 50)

# Encontrar la mejor k de la opción 1
best_k = max(best_k_results.keys(), key=lambda k: best_k_results[k]['r2'])
print(f"\n🏆 Usando las mejores {best_k} características + estandarización:")

# Aplicar selección de características
selector_best = SelectKBest(score_func=f_regression, k=best_k)
X_train_selected_best = selector_best.fit_transform(diabetes_data_train, diabetes_target_train)
X_test_selected_best = selector_best.transform(diabetes_data_test)

# Aplicar estandarización a las características seleccionadas
scaler_combined = StandardScaler()
X_train_combined = scaler_combined.fit_transform(X_train_selected_best)
X_test_combined = scaler_combined.transform(X_test_selected_best)

# Entrenar modelo combinado
model_combined = LinearRegression()
model_combined.fit(X_train_combined, diabetes_target_train)

# Hacer predicciones
pred_combined = model_combined.predict(X_test_combined)

# Calcular métricas
r2_combined = r2_score(diabetes_target_test, pred_combined)
mae_combined = mean_absolute_error(diabetes_target_test, pred_combined)
mse_combined = mean_squared_error(diabetes_target_test, pred_combined)

selected_features_best = np.array(diabetes.feature_names)[selector_best.get_support()]
print(f"• Características utilizadas: {', '.join(selected_features_best)}")
print(f"• R²: {r2_combined:.4f} (cambio: {r2_combined-base_r2:+.4f})")
print(f"• MAE: {mae_combined:.3f} (cambio: {mae_combined-base_mae:+.3f})")
print(f"• MSE: {mse_combined:.3f} (cambio: {mse_combined-base_mse:+.3f})")

# COMPARACIÓN FINAL DE TODOS LOS MODELOS
print(f"\n" + "=" * 60)
print("📊 COMPARACIÓN FINAL DE TODOS LOS MODELOS")
print("=" * 60)

models_comparison = {
    'Modelo Base (10 features)': {'r2': base_r2, 'mae': base_mae, 'mse': base_mse},
    f'Feature Selection ({best_k} features)': best_k_results[best_k],
    'Estandarización': {'r2': r2_scaled, 'mae': mae_scaled, 'mse': mse_scaled},
    'Combinado (FS + Escala)': {'r2': r2_combined, 'mae': mae_combined, 'mse': mse_combined}
}

print(f"{'Modelo':<25} {'R²':<8} {'MAE':<8} {'MSE':<8}")
print("-" * 50)
for model_name, metrics in models_comparison.items():
    print(f"{model_name:<25} {metrics['r2']:<8.4f} {metrics['mae']:<8.3f} {metrics['mse']:<8.3f}")

# Identificar el mejor modelo
best_model_name = max(models_comparison.keys(), key=lambda x: models_comparison[x]['r2'])
best_metrics = models_comparison[best_model_name]

print(f"\n🏆 MEJOR MODELO: {best_model_name}")
print(f"   • R²: {best_metrics['r2']:.4f}")
print(f"   • MAE: {best_metrics['mae']:.3f}")
print(f"   • MSE: {best_metrics['mse']:.3f}")

# Análisis de mejoras
improvement_r2 = best_metrics['r2'] - base_r2
improvement_mae = base_mae - best_metrics['mae']  # Reducción es mejor
improvement_mse = base_mse - best_metrics['mse']  # Reducción es mejor

print(f"\n📈 MEJORAS OBTENIDAS:")
print(f"   • Mejora en R²: {improvement_r2:+.4f} ({improvement_r2/base_r2*100:+.1f}%)")
print(f"   • Reducción en MAE: {improvement_mae:+.3f} ({improvement_mae/base_mae*100:+.1f}%)")
print(f"   • Reducción en MSE: {improvement_mse:+.3f} ({improvement_mse/base_mse*100:+.1f}%)")

# Recomendaciones finales
print(f"\n" + "=" * 60)
print("💡 RECOMENDACIONES Y CONCLUSIONES:")
print("=" * 60)

if improvement_r2 > 0.01:
    print("✅ Se lograron mejoras significativas en el modelo")
else:
    print("⚠️  Las mejoras fueron marginales")

print(f"\n📋 LECCIONES APRENDIDAS:")
print("• Feature Selection puede mejorar el rendimiento eliminando ruido")
print("• La estandarización es importante para algoritmos basados en distancia")
print("• La combinación de técnicas no siempre es la mejor opción")
print("• Es importante validar con múltiples métricas")

if best_model_name != 'Modelo Base (10 features)':
    print(f"\n🎯 RECOMENDACIÓN: Usar el {best_model_name} para futuros análisis")
else:
    print(f"\n🎯 RECOMENDACIÓN: El modelo base es suficiente para este dataset")

print("=" * 60)
````
